**Давайте добавим новый признак.**

```python
def age_category(age):
    '''
    < 30 -> 1
    >= 30, <55 -> 2
    >= 55 -> 3
    '''
    if age < 30:
        return 1
    elif age < 55:
        return 2
    elif age >= 55:
        return 3
```

```python
age_categories = [age_category(age) for age in data.Age]
data['Age_category'] = age_categories
```

**Другой способ сделать это с `apply`.**

```python
data['Age_category'] = data['Age'].apply(age_category)
```

** 4. Правда ли, что средний возраст выживших людей выше, чем у пассажиров, которые в конечном итоге умерли? **

```python
young = data[data['Age'] < 30]
old = data[data['Age'] > 60]
young_survived = young[young['Survived'] == 1]
old_survived = old[old['Survived'] == 1]

print("Молодые:", round(len(young_survived)/len(young)*100, 1), "%")
print("Пожилые:", round(len(old_survived)/len(old)*100, 1), "%")
```
